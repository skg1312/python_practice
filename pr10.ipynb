```python
sq=[]
for i in range(1,11):
    sq.append(i**2)
sq
```

```python
sq2 = [(i**2) for i in range(1,21)]
sq2
```

```python
sqev = []
for i in range(1,11):
    if i % 2 ==0:
        sqev.append(i**2)
sqev
```

```python
sqev2 = [i**2 for i in range(1,21) if i %2==0]
sqev2
```

```python
st = tuple(i**2 for i in range(1,21))
st
```

```python
sqev3 = [i**2 if i%2==0 else i**3 for i in range(1,21)]
sqev3
```

```python
sq3 = [i*j for i in range(1,20) for j in range(20,40)]
sq3
```
